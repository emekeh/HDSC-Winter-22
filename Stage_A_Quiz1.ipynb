### **Question 1:** Which of these Areas had the highest sum in 2017?

```python
# Area with the highest sum in 2017
df.groupby("Area")[["Y2017"]].sum().sort_values(by="Y2017", ascending=False).iloc[0]
```

### **Question 2:** What is the **mean** and **standard deviation** across the whole dataset for the year 2017 to 2 decimal places?

```python
# Mean and standard deviation across the whole dataset in 2017.
mean = df["Y2017"].mean()

std = df["Y2017"].std()

print(f"Mean for Year 2017: {round(mean, 3)}")

print(f"Standard Deviation for Year 2017: {round(std, 3)}")
```

### **Question 3:** What is the total number of the sum of `"Processing"` in 2017?

```python
# total number of the sum of Processing in 2017
proc_2017 = df.groupby("Element").sum().loc["Processing"]["Y2017"]
print(f'Sum of Processing in 2017 is {round(proc_2017, 2)}')
```

### **Question 11:** Which of these Areas had the 7th lowest sum in 2017?

```python
# Area with the 7th lowest sum in 2017
df.groupby("Area")[["Y2017"]].sum().sort_values(by="Y2017", ascending=False).iloc[-7]
```

### **Question 12:** What is the total sum of `"Wine"` produced in 2015 and 2018 respectively?

```python
# Total sum of Wine produced in 2015 and 2018 respectively
df.groupby("Item")[["Y2015", "Y2018"]].sum()
```

```python
# Total Sum of Wine produced in 2015 and 2018
wine_2015 = df.groupby("Item")[["Y2015", "Y2018"]].sum().loc["Wine"][0]

wine_2018 = df.groupby("Item")[["Y2015", "Y2018"]].sum().loc["Wine"][1]

print(f"Total sum of Wine produced in 2015: {round(wine_2015, 2)}")

print(f"Total sum of Wine produced in 2018: {round(wine_2018, 2)}")
```

### **Question 15:** What is the total number and percentage of missing data in 2014 to 3 decimal places?

```python
# Total number and Percentage of missing data in 2014
null_2014 = df["Y2014"].isnull().sum()
total_2014 = len(df["Y2014"])
null_percent = (null_2014/total_2014) * 100

print(f"Total missing data: {null_2014} \nPercent of missing data: {round(null_percent, 3)}")
```
